```python
import pyodbc
import sys
import logging
from datetime import datetime
```

```python
logging.basicConfig(
    filename="hotel.log",
    level=logging.DEBUG,
    format="%(username)s - %(asctime)s - %(levelname)s - %(message)s",
    datefmt="%d-%m-%Y %H:%M:%S",
    filemode="a",
    force=True
)    
```

```python
class OptCourses:

    def __init__(self):
        self.L1 = ["HR", "Finance", "Marketing", "DS"]
        self.total_cost=0
        self.list_sem=['semester',"anuual","no"]
        self.fees=200000
        self.analytics_fee=self.fees+self.fees*0.10
        self.hostel_rent=200000
        self.transporation_charges=13000
        self.food_per_month=2000  

    def Student_Signin(self):
        try:
            self.conn=pyodbc.connect(
                r'Driver=ODBC Driver 18 for SQL Server;'
                r'Server=SEJAL\SQLEXPRESS;'
                r'Database=Courses_DB;'
                r'Trusted_Connection=yes;')
            
        except:
            self.logger.error("Connection failed")
            print(sys.exc_info())


    def BasicSubjects(self):
        print(f"This are the options for selecting masters degree {self.L1}")
        while True:
            try:
                subject=input("Enter Subject:")
                if subject in self.L1:
                    print(f"You have selected {subject} field")
                    self.analytics=input("Do you have applied for analystics:(Y/N)").lower()
                
                    if self.analytics=='y':
                        print("You have opt for analytics")
                        self.facility=input("Do yo want some extra facilities (y/n):")
                        if self.facility=='y':
                            self.ExtraFacilities()
                        self.total_cost=self.fees+self.analytics_fee
                    else: # If analytics is 'n', just use base fees
                        self.facility=input("Do yo want some extra facilities (y/n):")
                        if self.facility=='y':
                            self.ExtraFacilities()
                        self.total_cost = self.fees


                else:
                    print("Please select valid course")
                break

            except ValueError():
                print("Please enter correct data")


    def ExtraFacilities(self):
        self.hostel=input("Do you have hostel?(Y/N)").lower()
        
        if  self.hostel=='y':
            self.total_cost += self.hostel_rent

        self.food=int(input("for How many months you have food facility)"))
        if self.food>0 :
            self.total_cost += self.food * self.food_per_month

        self.transporation=input("Do you have transportation?(semester/annual/no):").lower()
        if  self.transporation in self.list_sem and self.transporation != 'no': 
            self.total_cost += self.transporation_charges


        print("Your total fee is:",self.total_cost)

    
```
