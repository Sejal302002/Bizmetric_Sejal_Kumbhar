Comprehensive list:

```python
#1.Write a list comprehension to generate squares of numbers from 1 to 10

sqr=[(x,x*2) for x in range(1,10)]
sqr
```

```python
#2. Create a list of even numbers between 1 and 50 using list comprehension.
even_number=[ x for x in range(1,50) if x%2==0]
even_number
```

```python
#3. Convert all strings in a list to uppercase using list comprehension
a=["python","sql","database"]
uppercase_list=[x.upper() for x in a ]
uppercase_list
```

```python
#4. Given a list of integers, create a new list that contains only the positive numbers.

numbers=[1,0,-9,8,7,6,-4,-5]
positive_numbers=[x for x in numbers if x>0]
positive_numbers
```

```python
#5. Create a list of tuples (num, num^2) for numbers 1 to 5.

updated_list=[(num,num^2) for num in range(1,5)]
updated_list
```

```python
#6. Extract all vowels from a given string using list comprehension
str="Database"
vowels=[x for x  in str if x in "aeiou"]
vowels
```

```python
#7. Flatten a 2D list using list comprehension

arr = [[1, 2, 3],
          [4, 5, 6]]

flattened = [num for x in arr for num in x]
print(flattened)
```

```python
#8. Replace all negative numbers in a list with 0 using list comprehension.
numbers=[1,0,-9,8,7,6,-4,-5]
updated_list=[0 if x<0  else x for x in numbers]
updated_list
```

```python
#9. Given a list of words, create a list of lengths of each word.
a=["python","sql","database"]
length_list=[len(x) for x in a]
length_list
```

```python

#10. Filter out words that start with the letter 'A' or 'a'.

l1=['abcd','Ankita','sejal']
updated_list=[x for x in l1 if x[0]=='a' or x[0]=='A']
updated_list
```

```python
#11. From a list of numbers, generate a list of “even” or “odd” strings using list comprehension

numbers=[1,2,3,4,5,6,7,8]
even_odd=["even"if x%2==0 else "odd" for x in numbers ]
even_odd
```

```python
#12. Create a list of numbers divisible by both 3 and 5 in range 1–100.

div_numbers=[x for x in range(1,100) if x%3==0 and x%5==0]
div_numbers
```

```python
#13. Write a nested list comprehension to generate a multiplication table for 1–5
mul_table=[(x*i)for x in range(1,6)  for i in range(1,6)]
mul_table
```

```python
#14. Convert a dictionary’s keys into a list using list comprehension.
dict1={"roll":101,"name":"abc","std":"fy"}
fetching_keys=[x for x in dict1 if dict1.keys()]
fetching_keys
```

```python
#15. Extract numeric digits from a string using list comprehension.

str="abc123"
extract_digit=[x for x in str if x.isdigit()]
extract_digit
```

Lambda function

```python
#1. Write a lambda to add two numbers

add=lambda a,b:a+b
print(add(9,2))
```

```python
#2. Create a lambda to check if a number is even.

even_number=lambda a:a%2==0
even_number(4)
```

```python
#3. Write a lambda to get the last character of a string.
last_char=lambda x:x[-1]
last_char("sejal")
```

```python
#4. Use lambda with map() to square every number in a list.
nums = [1, 2, 3, 4, 5]

square = lambda x: x * x

result = list(map(square, nums))
result
```

```python
#5. Use lambda with filter() to get only odd numbers from a list.
nums = [1, 2, 3, 4, 5]
odd_nums=lambda x: x%2==1
result=list(filter(odd_nums,nums))
result
```

```python
#6. Use sorted() + lambda to sort a list of tuples by second value
nums = [(1, 5), (2, 3), (4, 1)]

result=sorted(nums,key=lambda x:x[1])
result
```

```python
#7. Create a lambda to check if a string is a palindrome
is_palindrome = lambda x: x == x[::-1]

is_palindrome("madam")

```
