```python

#17.

from datetime import datetime

fd_rates = {
    "public_existing": [
        (7, 45, 5.75),
        (46, 179, 6.25),
        (180, 210, 6.35),
        (211, 364, 6.40),
        (365, 729, 6.80),
        (730, 1094, 6.80),
        (1095, 1824, 6.80),
        (1825, 3650, 6.85)
    ],
    
    "public_revised": [
        (7, 45, 5.75),
        (46, 179, 6.25),
        (180, 210, 6.35),
        (211, 364, 6.40),
        (365, 729, 7.00),
        (730, 1094, 6.75),
        (1095, 1824, 6.70),
        (1825, 3650, 6.60)
    ],
    
    "senior_existing": [
        (7, 45, 6.25),
        (46, 179, 6.75),
        (180, 210, 6.85),
        (211, 364, 6.90),
        (365, 729, 7.30),
        (730, 1094, 7.30),
        (1095, 1824, 7.30),
        (1825, 3650, 7.35)
    ],
    
    "senior_revised": [
        (7, 45, 6.25),
        (46, 179, 6.75),
        (180, 210, 6.85),
        (211, 364, 6.90),
        (365, 729, 7.50),
        (730, 1094, 7.25),
        (1095, 1824, 7.20),
        (1825, 3650, 7.10)
    ]
}

def calculate_fd_interest():
    start_date = input("Enter FD Start Date (dd-mm-yyyy): ")
    end_date = input("Enter FD End Date (dd-mm-yyyy): ")

    start = datetime.strptime(start_date, "%d-%m-%Y")
    end = datetime.strptime(end_date, "%d-%m-%Y")

    tenure_days = (end - start).days
    print("Total FD Days:", tenure_days)

    customer = input("Enter customer type (public/senior): ").lower()
    rate_type = input("Enter rate type (existing/revised): ").lower()

    key = customer + "_" + rate_type

    if key not in fd_rates:
        print("Invalid option selected")
        return

    for min_days, max_days, rate in fd_rates[key]:
        if min_days <= tenure_days <= max_days:
            print(f"Interest Rate Applicable: {rate}%")
            return

    print("No matching FD bucket found.")
    

calculate_fd_interest()
```
