```python
import pyodbc

class Database:
    def __init__(self):
        self.conn = pyodbc.connect(
            r'Driver=ODBC Driver 18 for SQL Server;'
            r'Server=SEJAL\SQLEXPRESS;'
            r'Database=Appollo_DB;'
            r'Trusted_Connection=yes;'
            r'TrustServerCertificate=yes;'
        )
        self.cursor = self.conn.cursor()
        print("Database connected successfully!")
```

```python
class ApolloLogin():

    product_list = ['medicine', 'suppliments', 'injection']

    def __init__(self, db):
        self.db = db

        while True:
            have_acc = input("Do you have Apollo account (y/n): ").lower()

            if have_acc == 'y':
                self.login()
                break

            elif have_acc == 'n':
                self.register()
                break

            else:
                print("Invalid choice")

    
    def login(self):
        username = input("Enter username: ")
        password = input("Enter password: ")

        self.db.cursor.execute(
            "SELECT * FROM ApolloUsers WHERE Username=? AND Password=?",
            (username, password)
        )

        row = self.db.cursor.fetchone()

        if row:
            print("Welcome to Apollo")
            print("----------------------------------")
            print(f"Hello, {username}")
            print("Products are:", self.product_list)

            product = input("What product are you looking for: ").lower()

            if product in self.product_list:
                if product == 'medicine':
                    print("You are in medicine category")
                    print("--------------")
                    med = Medicines()
                    med.SearchMedicines()
                    print("--------------")
                else:
                    print("Feature coming soon.")
            else:
                print("Invalid product.")

        else:
            print("Invalid credentials.")

    
    def register(self):
        print("Create your Apollo account")

        username = input("Enter username: ")
        password = input("Enter password: ")

        try:
            self.db.cursor.execute(
                "INSERT INTO ApolloUsers (Username, Password) VALUES (?, ?)",
                (username, password)
            )
            self.db.conn.commit()

            print("Account created successfully!")
            print("Please login now.\n")

            self.login()

        except Exception as e:
            print("Registration failed:", e)
```
