```python
class ApolloLogin():
    product_list=['medicine','suppliments','injection']
    
    def __init__(self):
        while True:
            try:
                have_acc=input("Do you have Apollo account (y/n)").lower()
                if have_acc=='y':
                    self.login()
                    
                elif have_acc=='n':
                    self.register()
                
                else:
                    print("Invalid choice")
                break
            except ValueError:
                print("Please enter either y(yes) or n(no)")

    def login(self):
       while True:
          try:
            username=input("Enter a username:")
            password=input("Enter a password:")
            
            if username in login_data and login_data[username]==password:
                  print("Welcome to the Apollo")
                  print("----------------------------------")
                  print(f"Hello,{username}")
                  
                  print("Products are:",self.product_list)
                  product=input("What product are you looking for:").lower()

                  if product in self.product_list:
                      if product=='medicine':
                            print("You are in medicine category")
                            print("--------------")
                           
                            med = Medicines()
                            med.SearchMedicines()
                            print("--------------")

                      else:
                        print("please enter valid data")

                

            else:
               print("Please enter valid credentials")

            break 
          except ValueError:
             print("Please enter correct data") 
          
    def register(self):
       while True:
            try:
                print("Please enter details to create your account")
                username=input("Please enter your name to create account")
                password=input("Please set a password")
                
                if username.isalpha():
                    login_data[username]=password
                    print("Account created succesfully")
                    print("Please login now!")
                    self.login()
                
                else:
                    print("Account creation is failed.")

                break
            except ValueError:
                print("Invalid credentials")
          

          

    

       
            
         
```
